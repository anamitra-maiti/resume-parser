```python
import re
import string
def cleanResume(resumeText):
    resumeText = re.sub(r'http\S+\s*',' ',resumeText)  # Removes URl
    resumeText = re.sub(r'RT|cc',' ',resumeText) # Removes RT and cc
    resumeText = re.sub(r'#\S+','',resumeText) # Removes hashtags
    resumeText = re.sub(r'@\S+',' ',resumeText) # Removes mentions
    resumeText = re.sub(r'[%s]' % re.escape(string.punctuation), ' ', resumeText) # Removes punctuations
    resumeText = re.sub(r'[^\x00-\x7f]',r' ', resumeText) 
    resumeText = re.sub(r'\s+', ' ', resumeText)  # Removes extra whitespace
    return resumeText


df["cleaned_Resume"] = df.Resume.apply(lambda x: cleanResume(x))
    
```

```python
import nltk
nltk.download('punkt_tab')
  
from nltk.corpus import stopwords
oneSetOfStopWords = set(stopwords.words('english')+['``',"''"]) # Set of most commonly used stopwords in english language
totalWords =[]
Sentences = df['Resume'].values
cleanedSentence = ""
for records in Sentences:
    cleanedText = cleanResume(records)
    cleanedSentence += cleanedText
    requiredWords = nltk.word_tokenize(cleanedText)
    for word in requiredWords:
        if word not in oneSetOfStopWords and word not in string.punctuation:
            totalWords.append(word)
wordfreqdist = nltk.FreqDist(totalWords)
most_common = wordfreqdist.most_common(50)
print(most_common)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack

requiredText = df['cleaned_Resume'].values
requiredTarget = df['Category'].values
```

```python
word_vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    stop_words='english')
word_vectorizer.fit(requiredText)
WordFeatures = word_vectorizer.transform(requiredText)
print ("Feature completed .....")
X_train,X_test,y_train,y_test = train_test_split(WordFeatures,requiredTarget,random_state=42, test_size=0.2,
                                                 shuffle=True, stratify=requiredTarget)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.neighbors import KNeighborsClassifier

clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train, y_train)
prediction = clf.predict(X_test)
print('Accuracy of KNeighbors Classifier on training set: {:.2f}'.format(clf.score(X_train, y_train)))
print('Accuracy of KNeighbors Classifier on test set:     {:.2f}'.format(clf.score(X_test, y_test)))
```

```python
from sklearn.metrics import classification_report

print("\n Classification report for classifier %s:\n%s\n" % (clf, classification_report(y_train, clf.predict(X_train))))
```

```python
print("\n Classification report for classifier %s:\n%s\n" % (clf, classification_report(y_test, prediction)))
```
